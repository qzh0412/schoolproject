```python
def update_ants(bar, x, y):
    bar[x] = ' '
    bar[y] = ' '
    x += 1
    y -= 1
    print('\r', end='')
```

```python
def update_show_bar(bar):
    for ch in bar:
        print(ch, sep='', end='')
        
def setup_ants(bar,x,y):
    bar[x] = '*'
    bar[y] = '*'
    update_show_bar(bar)
    
```

```python
def setup_bar(bar,numlength):
    for i in range(0,numlength):
        bar.append(' ')
```

# list copy

```python
# 赋值浅copy 指向同一list
list_1 = [1, [2,3]]
list_2 = list_1
list_2[0] = 1.5
print(list_1)
print(list_2)

# 赋值后 另外指向一list 则不同
print('______________________________')
list_1 = [1,[2,3]]
list_2 = list_1
list_2 = [1.5,[2,3]]
print(list_1)
print(list_2)

##copy函数 不改变原list
print('______________________________')
list_1 = [1,[2,3]]
list_2 = list_1.copy()
list_2.append(4)
print(list_1)
print(list_2)

#改变同指向 改变原list
print('______________________________')
list_2[1][0] = 250
print(list_1)
print(list_2)
```

# 参数不同形式

```python
def print_params(*params):
    for element in params:
        print(element, end=' ')
    print()
    
print_params(1)
print_params(1,2,3)
print_params(1,2,3,4,5)
```

```python
def sum_params(*params):
    result = 0
    for element in params:
        result += element
    return result

print(sum_params(1))
print(sum_params(1,2,3))
```

# 非默认缺省顺序参数

```python
def interval(start, stop=None, step=1):
    result = []
    
    if stop is None:
        start, stop = 0, start
    
    i = start
    while i < stop:
        result.append(i)
        i += step
    return result

print(interval(0, 10, 2))
print(interval(0, 5, 1))
print(interval(4))
```

# 函数嵌套

```python
def multiplier(factor):
    def multiplyByFactor(number):
        return number * factor 
    return multiplyByFactor

double = multiplier(2)
print(double(4))
print('______________________________')
triple = multiplier(3)
print(triple(4))
```

# 迭代

```python
def Fib(i):
    if i == 1 or i== 2 :
        return 1
    else:
        return Fib(i - 1) + Fib(i - 2)

Fib(5)        
```

```python
def Factorial(i):
    if i == 1 or i == 0:
        return 1
    else:
        return i * Factorial(i-1)
    
Factorial(5)
```

```python
# 叠64层塔
# 第1~n-1层 a->b  第n层 a->c 第1~n-1层 b->c

def hanoi(n, source, dest, temp):
    if n == 1 :
        print('{}->{}'.format(source, dest))
    else :
        hanoi(n-1, source, temp, dest)
        print('{}->{}'.format(source, dest))
        hanoi(n-1, temp, dest, source)

hanoi(2, 'a', 'c', 'b')
hanoi(64, 'a', 'c', 'b')
```
